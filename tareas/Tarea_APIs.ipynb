```python
buscar = 'ciencia'
endpoint = 'https://en.wikipedia.org/w/api.php'
params = {
    'action':'query',
    'format':'json',
    'list':'search',
    'srsearch': buscar,
    'prop':'extracts',
    'srlimit':20
            }
wiki_req = requests.get(endpoint, params=params)
print(wiki_req.status_code)

datos = wiki_req.json()['query']['search']
res = []
cont = []

for result in datos:
    params = {
        'action': 'query',
        'format': 'json',
        'prop': 'extracts',
        'exintro': True,
        'explaintext': True,
        'pageids': result['pageid']
    }
    wiki_req = requests.get(endpoint, params=params)
    result = wiki_req.json()['query'].get('pages', {})
    resumen = list(result.values())[0].get('extract', '')

    cont_fisica = int('fisica' in resumen.lower())
    cont_ciencia = int('astro' in resumen.lower())

    res.append(resumen)
    cont.append(cont_fisica or cont_ciencia)

dataframe = pd.DataFrame({
    'pageid': [result['pageid'] for result in datos],
    'titulo': [result['title'] for result in datos],
    'resumen': res,
    'con_palabra': cont
})

dataframe
```
